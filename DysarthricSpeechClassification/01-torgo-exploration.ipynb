# Class distribution

```python
labels = torgo.labels

# Get class distribution
classes, counts = np.unique(labels, return_counts=True)

# Optional: map labels to human-readable names
label_names = ['Non-Dysarthric', 'Dysarthric']  # assuming 0 = Control, 1 = Dysarthric

# Custom function to show both percentage and count
def autopct_format(pct, allvals):
    absolute = int(np.round(pct / 100. * np.sum(allvals)))
    return f"{pct:.1f}%\n({absolute})"

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    counts,
    labels=label_names,
    autopct=lambda pct: autopct_format(pct, counts),
    startangle=90,
    colors=['skyblue', 'salmon']
)
plt.title('TORGO Dataset Class Distribution')
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```
